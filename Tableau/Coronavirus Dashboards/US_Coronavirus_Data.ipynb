```python
import pandas as pd
import os
```

#### CASES

```python
us_cases = pd.read_csv('time_series_covid19_confirmed_US.csv')
```

```python
us_cases
```

```python
states_c = list(us_cases['Province_State'].unique())
```

```python
case_totals=[]
for state in states_c:
    new_df = pd.DataFrame()
    new_df = us_cases[us_cases['Province_State']==state]
    day_cases=[state]
    for i in range(11,len(list(us_cases.columns))):
        day_cases.append(new_df.iloc[:,i].sum())
    
    case_totals.append(day_cases)
```

```python
dates = list(us_cases.columns)[11::]
dates.insert(0, 'State')
```

```python
cases_ts = pd.DataFrame(case_totals, columns=dates)
```

```python
cases_ts.head()
```

```python
cases_ts.drop([56,57])
```

```python
cases_ts.to_csv('state_cases_ts.csv', index=False)
```

```python
cases_total = cases_ts[['State', '7/18/20']]
```

```python
cases_total = cases_total.rename(columns={'7/18/20': 'Total cases'})
```

```python
cases_total.head()
```

```python
cases_total.to_csv('states_cases_totals_0718.csv', index=False)
```

#### DEATHS

```python
us_deaths = pd.read_csv('time_series_covid19_deaths_US.csv')
```

```python
states_d = list(us_deaths['Province_State'].unique())
```

```python
death_totals=[]
for state in states_d:
    new_df = pd.DataFrame()
    new_df = us_deaths[us_deaths['Province_State']==state]
    day_deaths=[state]
    for i in range(12,len(list(us_deaths.columns))):
        day_deaths.append(new_df.iloc[:,i].sum())
    
    death_totals.append(day_deaths)
```

```python
dates = list(us_deaths.columns)[12::]
```

```python
dates.insert(0, 'State')
```

```python
deaths_ts = pd.DataFrame(death_totals, columns=dates)
```

```python
deaths_ts = deaths_ts.drop([56,57])
```

```python
deaths_ts.to_csv('state_deaths_ts.csv', index=False)
```

```python
deaths_total = deaths_ts[['State', '7/18/20']]
```

```python
deaths_total = deaths_total.rename(columns={'7/18/20': 'Total deaths'})
```

```python
deaths_total.head()
```

```python
deaths_total.to_csv('states_deaths_totals_0718.csv', index=False)
```

#### COMBINED

```python
combined_df = pd.concat([cases_total, deaths_total['Total deaths']], axis=1)
```

```python
combined_df.head()
```

```python
combined_df['Mortality rate'] = combined_df['Total deaths'] / combined_df['Total cases']    
```

```python
combined_df.head()
```

```python
combined_df.to_csv('cases_deaths_0718.csv', index=False)
```
